```python
import pandas as pd
import sqlite3
from patient import Patient
from doctor import Doctor
from ward import Ward
from appointment import Appointment

conn = sqlite3.connect('Patient.db')
cursor = conn.cursor()
filename = "G03_Hospital_Patients.csv"
def load_patients_from_csv(csv_file):
    df_patient = pd.read_csv(csv_file,delimiter=';')
    patients = []
    
    for index, row in df_patient.iterrows():
        patient = Patient(
            row['patient_id'],
            row['doctor_name'], #Colunas no ficheiro estão trocadas
            row['age'],
            row['diagnosis'],
            row['condition'],
            row['wards_id']
        )
        patients.append(patient)
    return patients

def load_doctors_from_csv(csv_file):
    df_doctor = pd.read_csv(csv_file,delimiter=';')
    doctors = []
    
    for index, row in df_doctor.iterrows():
        doctor = Doctor(
            row['doctor_id'],
            row['patient_name'], #Colunas no ficheiro estão trocadas
            row['specialization']
        )
        doctors.append(doctor)
    return doctors

def load_wards_from_csv(csv_file):
    df_ward = pd.read_csv(csv_file,delimiter=';')
    wards = []
    
    for index, row in df_ward.iterrows():
        ward = Ward(
            row['wards_id'],
            row['ward_name'],
            row['floor']
        )
        wards.append(ward)
    return wards

def load_appointments_from_csv(csv_file):
    df_appointment = pd.read_csv(csv_file,delimiter=';')
    appointments = []
    
    for index, row in df_appointment.iterrows():
        appointment = Appointment(
            row['appointment_date'],
            row['doctor_id'],#Colunas no ficheiro estão trocadas
            row['patient_id'],#Colunas no ficheiro estão trocadas
            row['wards_id']
        )
        appointments.append(appointment)
    return appointments

patients = load_patients_from_csv(filename)
unique_patients = {patient.id: patient for patient in patients}.values()
unique_patients = list(unique_patients) 
#for patient in unique_patients:
    #patient.save_to_db(cursor)

doctors = load_doctors_from_csv(filename)
unique_doctors = {doctor.id: doctor for doctor in doctors}.values()
unique_doctors = list(unique_doctors) 
#for doctor in unique_doctors:
    #doctor.save_to_db(cursor)

wards = load_wards_from_csv(filename)
unique_wards = {ward.id: ward for ward in wards}.values()
unique_wards = list(unique_wards) 
#for ward in unique_wards:
    #ward.save_to_db(cursor)

appointments = load_appointments_from_csv(filename)
for appointment in appointments:
    appointment.save_to_db(cursor)


print("\nPatients:")
for row in cursor.execute("SELECT * FROM Patient;"):
    print(row)

print("\nDoctors:")
for row in cursor.execute("SELECT * FROM Doctor;"):
    print(row)

print("\nWards:")
for row in cursor.execute("SELECT * FROM Ward;"):
    print(row)

print("\nAppointments:")
for row in cursor.execute("SELECT * FROM Appointment;"):
    print(row)


conn.close()
```
